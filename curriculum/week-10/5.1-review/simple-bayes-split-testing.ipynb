You work for a company. 

You're running an A/B test on the color of your product. Red vs. Green. On front page, you have a pic of your product - a Gift Box. 

You measure how many people actually purchase the product out of how many people arrive at the site.


```python
historical_purchases = 25
historical_failures = 75
```

```python
red_rate = 0.31
green_rate = 0.19
n_red = 200
n_green = 200
```

```python
import scipy.stats as stats
```

```python
red_bern = stats.bernoulli(red_rate)
green_bern = stats.bernoulli(green_rate)
```

```python
red_obs = red_bern.rvs(n_red)
green_obs = green_bern.rvs(n_green)
```

```python
print np.mean(red_obs), np.mean(green_obs)
```

```python
stats.ttest_ind(red_obs, green_obs)
```

```python
red_beta = stats.beta(np.sum(red_obs)+1+historical_purchases, 
                      len(red_obs)-np.sum(red_obs)+1+historical_failures)
green_beta = stats.beta(np.sum(green_obs)+1+historical_purchases, 
                        len(green_obs)-np.sum(green_obs)+1+historical_failures)
```

```python
purchase_rates = np.linspace(0.001, 0.999, 200)
```

```python
fig, ax = plt.subplots(figsize=(7,5))
red_likelihood = red_beta.pdf(purchase_rates)
green_likelihood = green_beta.pdf(purchase_rates)
prior_like = prior_rate.pdf(purchase_rates)

ax.plot(purchase_rates, red_likelihood, color='darkred', lw=3)
ax.plot(purchase_rates, green_likelihood, color='green', lw=3)
ax.plot(purchase_rates, prior_like, color='grey', alpha=0.5, lw=3)

plt.show()
```
